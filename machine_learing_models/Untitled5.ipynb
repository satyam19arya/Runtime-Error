```python
import seaborn as sns
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers
from sklearn.preprocessing import MinMaxScaler
```

```python

df=pd.read_csv('/content/drive/My Drive/Wellbeing_and_lifestyle_data_Kaggle.csv')
```

```python
df.describe()
```

```python
df.columns
```

```python
dff=df.drop(['LOST_VACATION',"DAILY_SHOUTING","BMI_RANGE","WEEKLY_MEDITATION","FLOW",'TODO_COMPLETED','DONATION'],axis=1)
```

```python
abs=pd.get_dummies(dff['AGE'])
```

```python
abs
```

```python
abq=pd.get_dummies(dff['GENDER'])
```

```python
dff=pd.concat([dff,abq,abs],axis=1)
```

```python
dff.drop(['DAILY_STRESS'],axis=1,inplace=True)
```

```python
X=dff.drop('WORK_LIFE_BALANCE_SCORE',axis=1)
Y=dff['WORK_LIFE_BALANCE_SCORE']
```

```python
X
```

```python

X
```

```python
normalizer = tf.keras.layers.Normalization(axis=-1)
```

```python
model=tf.keras.Sequential([
    layers.Dense(32, input_shape=(18,), kernel_initializer='normal', activation='relu'),
    layers.Dense(16, kernel_initializer='normal', activation='relu'),
    layers.Dense(8, kernel_initializer='normal', activation='relu'),
    layers.Dense(units=1)
])
```

```python
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
X.drop('DAILY_STRESS',axis=1,inplace=True)
```

```python
model.fit(X,Y,epochs=20,batch_size=32)
```

```python
model.predict([[3	,2,	5	,0	,5	,2	,5,	0,	7,	1,	4,	0,	1,	0,	0,	1	,0	,0]])
```

```python
model.save('wlf.h5')
```
